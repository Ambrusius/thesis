```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the data
sample = pd.read_parquet('HZeeg_ggF_MC_reduced_Zmodel_eepairs_28August_131variables_300events.parquet')
```

```python
print(sample.shape)
```

```python
#mass = np.sqrt(2*sample_small['el_pt'][z][i]*sample_small['el_pt'][z][j]*(np.cosh(sample_small['el_eta'][z][i]-sample_small['el_eta'][z][j]) - np.cos(sample_small['el_phi'][z][i]-sample_small['el_phi'][z][j])))/1000.

mass_array = np.sqrt(2*sample['el1_pt'].values*sample['el2_pt'].values*(np.cosh(sample['el1_eta'].values-sample['el2_eta'].values) - np.cos(sample['el1_phi'].values-sample['el2_phi'].values)))/1000.

fig, ax = plt.subplots(figsize=(8, 6))
ax.set_xlim(-10, 200)
ax2 = ax.twinx()
ax.hist(mass_array, bins=300, histtype='step', linewidth=2, label='All events')
ax2.hist(mass_array[sample['isZ']==1], bins=100, histtype='step',color='r' , linewidth=2, label='Z events')
ax.set_xlabel('Mass [GeV]')
ax.set_ylabel('All events')
ax2.set_ylabel('Z events')


fig, ax = plt.subplots(figsize=(8, 6))
ax.set_xlim(50, 130)
ax.set_ylim(0, 1000)
ax.hist(mass_array[sample['isZ']==0], bins=300, histtype='step', linewidth=2, label='non-Z events')
ax.hist(mass_array, bins=300, histtype='step' , linewidth=2, label='All events')
ax.set_xlabel('Mass [GeV]')
ax.legend()

print('no')

```
